```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import tensorflow as tf
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
bank_df = pd.read_csv('./Churn_Modelling.csv')
```

```python
bank_df.head()
```

```python
bank_df.drop(columns=['RowNumber','CustomerId','Surname'],inplace=True)
```

```python
bank_df.head(10)
```

```python
lb=LabelEncoder()
bank_df['Gender']=lb.fit_transform(bank_df['Gender'])
```

```python
bank_df.head(10)
bank_df['Geography'][9998]
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder()
```

```python
encoder_df = pd.DataFrame(ohe.fit_transform(bank_df[['Geography']]).toarray())
```

```python
pd.DataFrame(ohe.fit_transform(bank_df[['Geography']]).toarray())
```

```python
bank_df=bank_df.join(encoder_df)
```

```python
bank_df.drop(columns=['Geography'],inplace=True)
```

```python
bank_df
```

```python
# France Germany Spain


bank_df.rename(columns={0:'France',1:'Germany',2:'Spain'},inplace=True)
bank_df.head()
```

```python
y=bank_df['Exited']
```

```python
bank_df.drop(columns=['Exited'],inplace=True)
```

```python
X=bank_df
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
```

```python
X_test=sc.transform(X_test)
```

```python
X_train
```

```python
y_train=np.array(y_train)
y_test=np.array(y_test)
```

```python
ann = tf.keras.models.Sequential()
```

```python
ann.add(tf.keras.layers.Dense(6, activation='relu'))
ann.add(tf.keras.layers.Dense(6, activation='relu'))
ann.add(tf.keras.layers.Dense(1, activation='sigmoid'))
```

```python
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
ann.fit(X_train,y_train,batch_size=32,epochs=100)
```

```python
ann.summary()
```

```python
y_preds=ann.predict(X_test)
```

```python
y_preds
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_preds.shape
```

```python
y_preds=y_preds.flatten()
```

```python

y_preds=(y_preds>0.5).astype(int)
y_preds
```

```python
print(accuracy_score(y_test,y_preds)*100)
```

```python
print(confusion_matrix(y_test,y_preds))
```
